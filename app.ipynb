```python
import pandas as pd
import numpy as np
```

```python
f=pd.read_csv('Placement_data_full_class.csv')
```

```python
df = f.rename(columns={'ssc_p': '10th %', 'ssc_b': '10th Board', 'hsc_p': '12th %', 'hsc_b': '12th Board', 'hsc_s': '12th Stream', 'degree_p': 'Degree %', 'degree_t': 'Degree Type', 'workex': 'Work Experience', 'etest_p': 'Employability Test %', 'mba_p': 'MBA %',})
```

```python
df.drop(columns=['gender', 'sl_no','salary'], inplace=True)
```

```python
df['12th Board'].unique()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
cat_cols = df.select_dtypes(include=['object']).columns
en = LabelEncoder()
for col in cat_cols:
    df[col] = en.fit_transform(df[col])
```

```python
X=df.drop('status',axis=1)
y=df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=150, n_jobs=-1, random_state=42)
rf.fit(X_train,y_train)
```

```python
predictions = rf.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error, accuracy_score
r2 = r2_score(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy:.2f}")


print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
import joblib
joblib.dump(rf, 'placement.pkl')
```
